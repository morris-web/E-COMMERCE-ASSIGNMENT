***Performing Exploratory Data Analysis on E-Commerce Data***

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv(fr"../ecommerce_data_2k25.csv")  # Reading dataset from CSV file
```

```python
sns.countplot(data=df, x='purchased')
plt.title("Purchase Distribution")
plt.show()
```

```python
df['gender'].value_counts().plot.pie(autopct='%1.1f%%', title='Gender Distribution')
plt.show()
```

```python
df['age'].plot.hist(bins=10, title='Age Distribution')
plt.show()
```

```python
sns.boxplot(data=df, x='gender', y='session_time')
plt.title("Session Time by Gender")
plt.show()
```

```python
plt.figure(figsize=(13, 27))
df['country'].value_counts().plot.barh(title='Customers by Country')
plt.yticks(fontsize=7)  # Try values like 8, 10, 12
plt.show()
```

```python
df['purchased'].value_counts().plot.pie(labels=['No', 'Yes'], autopct='%1.1f%%', title='Purchase Rate')
plt.show()
```

```python
#Product category
plt.figure(figsize=(14,8))
df['product_category'].value_counts().head(10).plot.bar(title='Most Clicked Product Categories')
plt.show()
```

`Observation`:
Analysis of the current data reveals that a significantly higher number of sessions end without a purchase compared to those that do. This indicates a gap in the conversion funnel, presenting a valuable opportunity to optimize the customer journey and increase purchase rates.

`Goal`:
Enhance the overall shopping experience by identifying and addressing friction points in the user journey. The aim is to boost customer engagement, streamline decision-making, and ultimately drive higher conversion rates and revenue.
